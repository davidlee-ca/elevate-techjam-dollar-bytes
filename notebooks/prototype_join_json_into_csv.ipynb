```python
import pandas as pd
import json
import os
from pandas.io.json import json_normalize

customer_filename_prefix = 'customers/customers_'
customer_file = '00000.json'

with (open(customer_filename_prefix + customer_file)) as f:
    customer = json.load(f)['result']['customers']
```

```python
print(customer[0])
```

```python
p = pd.DataFrame(customer)
addrs = pd.io.json.json_normalize(p['addresses'])
df = p.drop('addresses', axis=1).join(addrs)
```
